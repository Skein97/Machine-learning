```python
import pandas as pd
data_frame = pd.read_csv('FIFA23_official_data.csv')
data_frame.shape
def to_float(x):
    repl_dict = {'[kK]': '*1e3', '[mM]': '*1e6', '[bB]': '*1e9', }
    return x.replace(repl_dict, regex=True).map(pd.eval)
        
```

```python

df1 = pd.DataFrame(data_frame, columns=['Name', 'Wage', 'Value'])
wage = df1['Wage'].replace({'€': '', 'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(float)
value = df1['Value'].replace({'€': '', 'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(float)
df1['Wage'] = wage
df1['Value'] = value
df1['Difference'] = df1['Value'] - df1['Wage']
df1.sort_values('Difference', ascending=False)
```

```python
import seaborn as sns
sns.set

graph = sns.scatterplot(x='Wage', y='Value', data=df1)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage,Value)", "(@Wage, @Value)"),
    ("Name", "@Name"),])

p = figure(title='FIFA 2023', x_axis_label='Wage', y_axis_label='Value', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('Wage', 'Value', size=10, source=df1)
show(p)
```
